```python
# 1. Copy the users and roles DataFrames from the examples above.
# Create the users DataFrame.

import pandas as pd
import numpy as np

users = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6],
    'name': ['bob', 'joe', 'sally', 'adam', 'jane', 'mike'],
    'role_id': [1, 2, 3, 3, np.nan, np.nan]
})
display(users)


# Create the roles DataFrame

roles = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['admin', 'author', 'reviewer', 'commenter']
})
roles

```

```python
# 2. What is the result of using a right join on the DataFrames?

pd.merge(users, roles, how='right', left_on='role_id', right_on='id', indicator=True)

# right join drops non matching rows from the users table. Keeps everything from roles even if it does not match
```

```python
# 3. What is the result of using an outer join on the DataFrames?

pd.merge(users, roles, how='outer', left_on='role_id', right_on='id', indicator=True)

# otter join includes all rows from both dataframes regardless if they match
```

```python
# 4. What happens if you drop the foreign keys from the DataFrames and try to merge them?

# users = users.drop(columns='role_id')
# roles = roles.drop(columns='id')
pd.merge(users, roles, how='outer', indicator=True)

# it looks similar to pd.concat. Both tables are now concat along axis=0.
```

```python
# 5. Join the employees and titles DataFrames together.

db = 'employees'

url = env.get_db_url(db)


```

```python
query = '''
select *
from employees
	join titles
		using (emp_no);
'''
```

```python
df = pd.read_sql(query, url)
```

```python
df.info()
```

```python
# 6. For each title, find the hire date of the employee that was hired most recently with that title.



df['hire_date'] = pd.to_datetime(df['hire_date'])


df.groupby(by='title').max().sort_values(by='hire_date', ascending=False)


# df[['title', 'hire_date']]
```
