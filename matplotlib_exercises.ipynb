# 2. Create and label 4 separate charts with the following equations (choose a range for x that makes sense):


y=âˆšx


y=x^3



y=2^x



y=1/(x+1)


## You can use functions from the math module to help implement some of the equations above.



```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math
```

```python
# y=âˆšx

import math 

x1 = range(1,21)
y1 = [math.sqrt(i) for i in x1]


plt.plot(x1, y1)
plt.title('y = âˆšx')
plt.xlabel('list of x')
plt.ylabel('âˆšx')
plt.show()

```

```python
# y=x^3


x2 = range(1,21)
y2 = [i ** 3 for i in x2]


plt.plot(x2, y2)
plt.title('$ y = x^3 $')
plt.xlabel('list of x')
plt.ylabel('x^3')
plt.show()

```

```python
# y=2^x

x3 = range(1,21)
y3 = [2**i for i in x3]
plt.plot(x3,y3)

plt.title('y = 2^x')
plt.xlabel('list of x')
plt.ylabel('2^x')
plt.show()
```

```python
# y=1/(x+1)


x4 = range(1,21)
y4 = [(1/(i+1)) for i in x4]
plt.plot(x4,y4)
plt.title('$ y=1/(x+1) $')
plt.xlabel('$x$')
plt.ylabel('$1/(x+1)$')
plt.show()


```

# 3. Combine the figures you created in the last step into one large figure with 4 subplots.


```python



plt.figure(figsize=(10,6))

plt.subplot(221)
plt.plot(x1,y1)

plt.subplot(222)
plt.plot(x2,y2)

plt.subplot(223)
plt.plot(x3,y3)

plt.subplot(224)
plt.plot(x4,y4)

plt.suptitle('matplotlib exercise')
plt.show()



```

```python


fig, ax = plt.subplots(2,2, sharey = True) #initialize at the top with the plt.subplots()

ax[0,0].plot(x1,y1) #call the ax variable and plot on it
ax[0,1].plot(x2,y2)
ax[1,0].plot(x3,y3)
ax[1,1].plot(x4,y4)

ax[0,0].set_xlabel('babies') #calling ax specific plot for individual label

fig.supxlabel('common x label') #calling the figure variable to get overall x-label
fig.supylabel('common y label')
fig.suptitle('my four plots')

plt.tight_layout()
plt.show()




```
