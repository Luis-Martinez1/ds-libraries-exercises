## Use seaborn's load_dataset function to load the iris database to answer the following questions:

```python
# pandas for tabular data manipulation
import pandas as pd

# numpy for vectorized matrix operations
import numpy as np

# matplotlib.pyplot for basic matplotlib functionality
import matplotlib.pyplot as plt

# seabon for more visualizations! 
import seaborn as sns

#load the iris dataset
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
# 1. What does the distribution of petal lengths look like?

#distribution of petal_length
sns.histplot(df.petal_length, bins=10)
plt.show()
```

```python
# 2. Is there a relationship between petal length and petal width?

#petal_length vs petal_width
sns.scatterplot(data=df, x='petal_length', y='petal_width')
plt.show()

# as petal length increases so does petal width
```

```python
# 3. Would it be reasonable to predict species based on sepal width and sepal length? 
#    For this, you'll visualize two numeric columns through the lense of a categorical column.

sns.scatterplot(data=df, x='sepal_length', y='sepal_width', hue='species')
plt.show()

# it may be resonable for some species, yes.
```

```python
# 4. Which features would be best used to predict species?

# note: try using a pairplot first. you dont have to plot variables one by one. 
#      you can plot multiple at the same time.

sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species')
plt.show()

# petal lenght and petal width would be best for predicting species.
```

```python
# 3. Load the swiss dataset from pydataset and read it's documentation. 
#    Create visualizations to answer the following questions:

data('swiss', show_doc=True)

df_swiss = data('swiss')
```

```python
df_swiss.sample(5)
```

```python
# Create an attribute named is_catholic that holds a boolean value of whether or not the province is Catholic. 
#  (Choose a cutoff point for what constitutes catholic)

cath_series = df_swiss.Catholic

is_catholic = cath_series > 50
df_swiss['is_catholic'] = is_catholic
```

```python
df_swiss.sample(5)
```

```python
# Does whether or not a province is Catholic influence fertility?

sns.barplot(data=df_swiss, y='Fertility', x='is_catholic')
plt.show()

# Being Catholic may influence fertility.
```

```python
# What measure correlates most strongly with fertility?
# you can use df.corr() it shows a chart of corellation coefficients
# also you can use a heatmap to see all variables and how they correlate. 


sns.barplot(data=df_swiss, y='Education', x='is_catholic')
plt.show()

# Education correlates strongly.
```

```python
# 4. Load the chipotle dataset from SQL, create a bar chart that shows the 4 most popular items and the 
#    revenue produced by each.

import env
db = 'chipotle'

url = env.get_db_url(db)

pd.read_sql('select * from orders limit 5', url)

```

```python
orders_df = pd.read_sql('select * from orders', url)
```

```python
orders_df.sample(5)
```

```python
# remove '$' from item_price
orders_df['item_price'] = orders_df['item_price'].str.replace('$', '').astype(float)
```

```python
most_pop=orders_df.groupby('item_name')[['quantity', 'item_price']].sum()
most_pop.head()
```

```python
top_items=most_pop.sort_values('quantity', ascending=False).head(4)
top_items
```

```python
top_items.reset_index()
```

```python
sns.barplot(data=top_items, y=top_items.index, x='item_price');
```

```python
# 5. Load the sleepstudy dataset from pydataset and read it's documentation. 
#    Use seaborn to create a line chart of all the individual subject's reaction times and a more prominant 
#    line showing the average change in reaction time.

data('sleepstudy', show_doc=True)

df_sleep = data('sleepstudy')
```

```python
df_sleep.sample(10)
```

```python
# you need to change subjects from integer. its treating subjects as a int gradient. 
# change subjects to strings, force it to teat as catagorical. you should have about 18 lines for each subject.

df_sleep.Subject = df_sleep.Subject.astype(str)
```

```python
# add line plot  for average reaction time right on top of orignial plot. two line plots can be coded right on
# top of eachother. 

sns.lineplot(data=df_sleep, y='Reaction', x='Days', hue='Subject')
sns.lineplot(data=df_sleep, y='Reaction', x='Days', c='black', linewidth=5)
plt.show()
```
